```python
import numpy as np
import pandas as pd
import seaborn as sns

# Definir tema
sns.set_theme(style='darkgrid')
```

```python
# Base de Dados
Base_Dados = sns.load_dataset('tips')

# Verificar
Base_Dados.head()
```

```python
# Traduzindo colunas
Base_Dados.rename( columns={
    'total_bill': 'Conta_Total',
    'tip': 'Gorjeta',
    'sex': 'Sexo',
    'smoker': 'Fumante',
    'day': 'Dia Semana',
    'time': 'Refeicao',
    'size': 'Pessoas da mesa'
}, inplace=True )
```

```python
# Grafico relplot
sns.relplot( x='Conta_Total', y='Gorjeta', data=Base_Dados );
```

```python
# Grafico Relplot
# Passando outro parametro de classe no grafico
sns.relplot( x='Conta_Total', y='Gorjeta', data=Base_Dados, hue='Sexo');
```

```python
# Grafico de Linha com 2 eixos
sns.relplot( x='Conta_Total', y='Gorjeta', data=Base_Dados, kind='line' );
```

```python
# Garfico de Linha com 2 eixos em cores distintas
sns.relplot( x='Conta_Total', y='Gorjeta', data=Base_Dados,
            kind='line', hue='Fumante' );
```

```python
sns.histplot( data=Base_Dados, x='Conta_Total');
```

```python
sns.histplot( data=Base_Dados, x='Conta_Total', hue='Fumante');
```

```python
sns.barplot( data=Base_Dados, x='Sexo', y='Conta_Total', hue='Fumante');
```

```python
sns.pairplot( Base_Dados );
```

```python
sns.pairplot(Base_Dados, hue='Sexo' );
```

```python
sns.pairplot(Base_Dados, kind='kde' );
```

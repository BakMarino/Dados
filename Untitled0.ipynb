```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Frases que vamos utilizar no exemplo
Frase = 'Lorem Ipsum é simplesmente uma simulação de texto da indústria tipográfica e de impressos, e vem sendo utilizado desde o século XVI, quando um impressor desconhecido pegou uma bandeja de tipos e os embaralhou para fazer um livro de modelos de tipos. Lorem Ipsum sobreviveu não só a cinco séculos, como também ao salto para a editoração eletrônica, permanecendo essencialmente inalterado. Se popularizou na década de 60, quando a Letraset lançou decalques contendo passagens de Lorem Ipsum, e mais recentemente quando passou a ser integrado a softwares de editoração eletrônica como Aldus PageMaker.'
```

```python
Frase
```

```python
nuvem_palavras = WordCloud(
    background_color='black',  # Cor de fundo
    width=1600,                # Largura da imagem
    height=800                 # Altura da imagem
).generate(Frase)
```

```python
fig, ax = plt.subplots( figsize=(10, 5) )
ax.imshow( nuvem_palavras, interpolation='bilinear' )
ax.axis('off')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Texto de exemplo
texto = "Python é uma linguagem de programação incrível. Python é fácil de aprender e poderosa."

# Criar uma instância do WordCloud
nuvem_palavras = WordCloud(
    background_color='black',  # Cor de fundo
    width=1600,                # Largura da imagem
    height=800                 # Altura da imagem
).generate(texto)              # Gerar a nuvem de palavras a partir do texto

# Exibir a nuvem de palavras
plt.figure(figsize=(10, 5))
plt.imshow(nuvem_palavras, interpolation='bilinear')
plt.axis('off')  # Remover eixos
plt.show()
```

```python
import pandas as pd
import plotly.express as px
```

```python
# Criando Dados

# Cidades
Cidade = [ 'Itapetininga', 'Angatuba', 'Sorocaba', 'Guarei',
           'Paranapanema', 'Alambari', 'Sarapui', 'São Miguel Arcanjo' ]

Estado = [ 'São Paulo','São Paulo','São Paulo','São Paulo',
          'São Paulo','São Paulo','São Paulo','São Paulo' ]

# Gerando a Latitude
Latitude = [ -23.5917, -23.4897, -23.5017, -23.3733,
            -23.3875, -23.5503, -23.6397, -23.8781 ]

# Gerando a Longitude
Longitude = [ -48.0475, -48.3167, -47.4667, -47.7833,
             -48.0000, -47.7833, -47.7833, -47.7833 ]

# Valor
Vendas = [ 100, 120, 90, 50, 70, 90, 250, 400 ]
```

```python
# Dicionario
Dicionario = {
    'Cidade' : Cidade,
    'Estado' : Estado,
    'Lat' : Latitude,
    'Lon' : Longitude,
    'Vendas' : Vendas
}

Base_Dados = pd.DataFrame( Dicionario )

Base_Dados
```

```python
# Mapa de calor Geografico

px.density_mapbox(
    Base_Dados,
    lat='Lat',
    lon='Lon',
    z='Vendas',
    radius=30,
    center=dict(lat=-23.5489, lon=-47.6388),
    zoom=8,
    mapbox_style='carto-positron'

)
```
